```python
import matplotlib.pyplot as plt
size=[2441,2312,1031,1233]
label=['A형', 'B형', 'O형', 'AB형']
plt.axis('equal') #원 그리기
plt.rc('font', family='NanumBarunGothic') 
plt.pie(size, labels=label)
plt.show()
```

```python
plt.axis('equal')
plt.rc('font', family='NanumBarunGothic') 
color=['green', 'yellow','hotpink', 'blue']
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0))
plt.legend()
plt.show()
```

```python
# 단계 3: 한글 폰트 설정
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm

# 마이너스 표시 문제
mpl.rcParams['axes.unicode_minus'] = False
	
# 한글 폰트 설정
path = '/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf'
font_name = fm.FontProperties(fname=path, size=18).get_name()
plt.rc('font', family=font_name)
fm._rebuild()
```

```python
import csv
f=open('age.csv')
data=csv.reader(f)
result=[]
name=input('인구구조가 알고 싶은 지역의 이름')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))


plt.style.use('ggplot')
plt.title(name+'지역의 인구구조')
plt.plot(result)   # 라인그래프
plt.show()
```

```python
import csv
f=open('age.csv')
data=csv.reader(f)
result=[]
for row in data:
  if '봉담' in row[0]:
    for i in row[3:]:
      value = int(i.replace(',', ''))
      print(value)
      result.append(int(value))

import matplotlib.pyplot as plt
plt.bar(range(101), result)   #막대그래프
plt.show()
```
